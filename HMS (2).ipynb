### Medical Record Class
A medical record contains 4 strings of different attributes
### Patient File
A patient file is a Linked list,... traversed by its last pointer and thus stores linearly and accessing in reverse manner

```python
class Record:
    def __init__(self):
        self.attributes = ["diagnosis","symptoms","prescription","allergies","nth visit"]
        self.data = {}
        for i in self.attributes:
            self.data[i]=None
        self.next = None
        self.prev = None
    def diagnose(self):
        self.data["diagnosis"] = input("Give in the diagnosis : ")
    def symptomize(self):
        self.data["symptoms"] = input("Give in the Symptoms : ")
    def prescribe(self):
        self.data["prescription"] = input("Give in the prescription : ")
    def allergize(self):
        self.data["allergies"] = input("Give in the Allergies found : ")
    def viewRecord(self):
        print(f"Record number {self.data["nth visit"]}")
        print("\tDiagnosis : " , self.data["diagnosis"])
        print("\tSymptoms : " , self.data["symptoms"])
        print("\tPrescription : " , self.data["prescription"])
        print("\tAllergies : " , self.data["allergies"])
        
class PFile:
    def __init__(self):
        self.latest_record = None
        self.record_count = 0
        
    def addRecord(self):
        if(self.latest_record !=None):
            self.latest_record.next = Record()
            self.latest_record.next.prev = self.latest_record
            self.latest_record = self.latest_record.next
            self.latest_record.data["nth visit"] = self.record_count+1
            self.record_count+=1
        else:
            self.latest_record = Record()
            self.latest_record.data["nth visit"] = self.record_count+1
            self.record_count = 1
    def updatelatestRecord(self):
        self.latest_record.diagnose()
        self.latest_record.symptomize()
        self.latest_record.prescribe()
        self.latest_record.allergize()
        
    def deletelatestRecord(self):
        self.latestRecord = None
        
    def viewfile(self):
        temp = self.latest_record
        while(temp!=None):
            temp.viewRecord()
            temp = temp.prev
```

## Patient Class

The class which integrates all the above mentioned classes and give a complete data instance of a Patient

```python
import datetime
class Patient():
    def __init__(self):
        self.attributes = ["status","firstname","lastname","dob","age","gender","blood_group","contact","email","address","emergency_contact","registered_date","is_active","medical_history","current_medications","insurance_details","disease"]
        self.__data = {}
        for i in self.attributes:
            self.__data[i]=None
        self.flag=False
    
    def addPatient(self,fname="Un",lname="known",gender=None,blood="O",dob="01/01/2007"):
        self.__data["firstname"]=fname
        self.__data["lastname"]=lname
        self.dobdate = datetime.datetime(int(dob[6:10]),int(dob[3:5]),int(dob[0:2]))
        self.today = datetime.datetime.now()
        self.__data["dob"]=dob
        self.__data["gender"]=gender
        self.__data["blood_group"]=blood
        self.__data["age"] = self.today.year - self.dobdate.year - ((self.today.month, self.today.day)<(self.dobdate.month, self.dobdate.day))
        self.__data["registered_date"]=self.today
        self.flag=True
        self.addContact()
        self.addHistory()

    def addContact(self,con="100",em="hms@gmail.com",add="Patient, This hospital",econ="108"):
        if self.flag:
            self.__data["contact"]=con
            self.__data["email"]=em
            self.__data["address"]=add
            self.__data["emergency_contact"]=econ
            return
        print("First Add Patient Details")
        print("-"*35)

    def addHistory(self,dis=None,lvd=None,stat="Out Patient",cmed=None,mhis=None,insdet="No Insurance"):
        if self.flag:
            if(mhis == None):
                self.__data["medical_history"] = PFile()
            if(dis == None):
                self.__data["disease"] = Diseases()
            self.__data["current_medications"]=cmed
            self.__data["status"]=stat
            self.__data["insurance_details"]=insdet
            self.__data["last_visited_doctor"]=lvd
            return
        print("First Add Patient Details")
        print("-"*35)
        
    def update(self,att,data):
        if att not in self.__data:
            print(f"{att}__ not found in data")
            print("-"*35)
            return
        if(att=="dob"):
            self.__data["dob"] = data
            self.__dobdate = datetime.datetime(int(self.__data["dob"][6:10]),int(self.__data["dob"][3:5]),int(self.__data["dob"][0:2]))
            self.__data["age"] = self.today.year - self.dobdate.year - ((self.today.month, self.today.day)<(self.dobdate.month, self.dobdate.day))
        print(f"{att} changed from {self.__data[att]} to {data}")
        print("-"*35)
        self.__data[att]=data

    def showAtt(self,att):
        if att not in self.__data:
            print(f"{att} not found in data")
            print("-"*35)
            return
        print(f"{att} : {self.__data[att]}")
        print("-"*35)

    def returnAtt(self,att):
        if att not in self.__data:
            raise(f"{att} not found in data")
            return
        return self.__data[att]       
        
    def show(self):
        for i in self.attributes:
            if(i == "medical_history"):
                self.__data["medical_history"].viewfile()
            elif(i == "disease"):
                self.__data["disease"].showDiseases()
            else:
                print(i,":",self.__data[i])
            print("-"*35)
```

## Doctor Class
Integrates all of the data in doctor, nurse and enumeration

```python
class Doctor():
    def __init__(self):
        self.attributes = ["status","firstname","lastname","gender","specialization","contact","email","schedule","assigned_nurses"]
        self.__data = {}
        for i in self.attributes:
            self.__data[i]=None
        self.__data["schedule"] = []
        self.flag=False

    def addDoctor(self,stat=False,fname="Un",lname="Known",gen=None,spec="General",con="100",em="doctor@gmail.com",sche=[],anur=[]):
        if(sche == []):
            self.__data["schedule"] = Schedule()
        self.__data["status"]=stat
        self.__data["firstname"]=fname
        self.__data["lastname"]=lname
        self.__data["gender"]=gen
        self.__data["specialization"]=spec
        self.__data["contact"]=con
        self.__data["email"]=em
        self.__data["assigned_nurses"]=anur

    def update(self,att,data):
        if att not in self.__data:
            print(f"{att} not found in data")
            print("-"*35)
            return
        if att=="schedule":
            self.__data["schedule"].updateSchedule()
        else:
            print(f"{att} changed from {self.__data[att]} to {data}")
            print("-"*35)
            self.__data[att]=data

    def showAtt(self,att):
        if att not in self.__data:
            print(f"{att} not found in data")
            print("-"*35)
            return
        if att=="schedule":
            self.__data["schedule"].viewSchedule()
        else:
            print(f"{att} : {self.__data[att]}")
            print("-"*35)

    def returnAtt(self,att):
        if att not in self.__data:
            raise(f"{att} not found in data")
            return
        return self.__data[att]
        
    def show(self):
        for i in self.attributes:
            if(i == "schedule"):
                self.__data["schedule"].viewSchedule()
            else:
                print(i,":",self.__data[i])
        print("-"*35)
```

## Random Hospital Generator
Used for generating the hospital data and layout

```python
import random as r

def createMatrix(m,n):
    d = []
    for i in range(m):
        
        d.append([])
        for j in range(n):
            d[i].append(" ")
    return d

def disp(m):
    for i in range(len(m)):
        for j in range(len(m[i])):
            if type(m[i][j])!=list:
                print(m[i][j],end=" ")
            else:
                if m[i][j][0]!=False:
                    print("D",end=" ")
                else:
                    print("N",end=" ")
        print()

def CreateHospital(A,m,n):
    if m<5 or n<5:
        return A
    if m<8 or n<8:
        for i in range(m):
            A[i][n//2]="+"
        A = ver(A,m,n,r.random())
        A = hor(A,m,n,r.random())
        return A
    if m<=n:
        forty = (m*4)//10
        #for j in range(0,m//2+1,min(5,max(3,forty//2))):
        for j in range(0,m-1,min(5,max(3,forty//2))):
            if j!=0:
                for i in range(n):
                    A[m-j-1][i]="+"
                for i in range(m):
                    A[i][n-j-1]="+"
    else:
        forty = (n*4)//10
        #for j in range(0,n//2+1,min(5,max(3,forty//2))):
        for j in range(0,n-1,min(5,max(3,forty//2))):
            if j!=0:
                for i in range(m):
                    A[i][n-j-1]="+"
                for i in range(n):
                    A[m-j-1][i]="+"
    A = segment(A,m//2,n//2," ")
    A = CreateHospital(A,m//2+1,n//2+1)
    A = border(A,m//2+1,n//2+1)
    A = ver(A,m,n,r.random())
    A = hor(A,m,n,r.random())
    if m==len(A) and n==len(A[0]):
        A = border(A)
        return corner(A,"+")
    return border(A)

def segment(M,m=None,n=None,v=" "):
    if m==None:
        m = len(M)
    if n==None:
        n = len(M[0])

    for i in range(m+1):
        for j in range(n+1):
            M[i][j]=v
    return M

def border(M,x=None,y=None):
    if y==None:
        y = len(M)-1
    if x==None:
        x = len(M[0])-1
    #b = '\u25a2'
    for i in range(y+1):
        M[i][x]="+"
        M[i][0]="+"
    for i in range(x+1):
        M[y][i]="+"
        M[0][i]="+"
    return M

def ver(A,m=None,n=None,p=1):
    if m==None:
        m = len(A)
    if n==None:
        n = len(A[0])

    if r.random()<=p:
        for i in range(m):
            a = A[i]
            A[i] = a[n::-1]+a[n+1:]
    return A

def tra(A):
    B = createMatrix(len(A[0]),len(A))
    for i in range(len(A)):
        for j in range(len(A[i])):
            B[j][i]=A[i][j]
    return B

def hor(A,m=None,n=None,p=1):
    if m==None:
        m = len(A)
    if n==None:
        n = len(A[0])

    B = tra(A)
    B = ver(B,n,m,p)
    return tra(B)

def corner(A,D):
    id=1
    for i in range(1,len(A)):
        for j in range(len(A[i])-1):
            if A[i-1][j]==D and A[i-1][j+1]==D and A[i][j+1]==D:
                if r.random()<0.3:
                    A[i][j]=[id,0,True] # [id,priority,empty]
                    id+=1
                else:
                    A[i][j]=[False,0,False] # [!available,priority,empty]
    return A

def compare(A,B):
    if len(A)!=len(B) or len(A[0])!=len(B[0]):
        print("Size not equal, So technically They are 100% not Equal")
        return False
    count=0
    for i in range(len(A)):
        for j in range(len(A[i])):
            if A[i][j]!=B[i][j]:
                count+=1
    temp = len(A)*len(A[0])
    print("Comparison complete")
    print(f"Both A and B are {round((temp-count)/temp*100,2)}% Alike")
    if count==0:
        return True
    return False


s = 15
print(f"Creating a Hospital with size {s} x {s}")
disp(CreateHospital(createMatrix(s,s),s,s))
```

## Hospital Class
Incorporation of all other classes - Patient, Doctor, Appointment, Bills

```python
import random as r

class Hospital():
    def __init__(self):
        self.names=["Vikram","Bhuvan","Sachcith","Ranjith","Advaith","Jashwanth","Raguram",
                    "Rahul","Kabilan","Kishore","Naveen","Subash","Cibi","Kushal","Charan",
                    "Haldar","Paul","Prajwal","Manas","Doctor","Patient"]
        self.special=["Cardio","Neurology","Pedio","Physio"]
        self.relation=["Father","Brother","Uncle","Friend","GrandFather","Cousin","Son","GrandSon","Nephew"]
        self.__doctor={}
        self.__patient={}
        self.appointment={}
        self.bill={}

    def HospitalCreater(self,s):
        self.__hospital = CreateHospital(createMatrix(s,s),s,s)
        
    def countDoctor(self):
        count=0
        for i in range(len(self.__hospital)):
            for j in range(len(self.__hospital)):
                if type(self.__hospital[i][j])==list and self.__hospital[i][j][0]!=False:
                    count+=1
        return count

    def showHospital(self):
        disp(self.__hospital)

    def blood(self):
        temp = r.random()
        if temp<0.35:
            return "O+"
        elif temp<0.48:
            return "O-"
        elif temp<0.78:
            return "A+"
        elif temp<0.86:
            return "A-"
        elif temp<0.88:
            return "B+"
        elif temp<0.96:
            return "B-"
        elif temp<0.99:
            return "AB+"
        else:
            return "AB-"

    def createDoctors(self):
        max_id = self.countDoctor()
        for i in range(1,max_id+1):
            self.__doctor[i]=Doctor()
            temp = self.names[r.randint(0,len(self.names)-1)]
            spec = self.special[r.randint(0,len(self.special)-1)]
            phone = r.randint(6000000000,9999999999)
            email="Doctor."+temp+"@gmail.com"
            self.__doctor[i].addDoctor(True,"Dr",temp,"M",spec,phone,email)

    def createPatients(self):
        max_pid = 2*self.countDoctor()
        for i in range(1,max_pid+1):
            self.__patient[i]=Patient()
            temp1 = self.names[r.randint(0,len(self.names)-1)]
            blood = self.blood()
            self.__patient[i].addPatient("Mr",temp1,"M",blood)
            phone1 = r.randint(6000000000,9999999999)
            email1 = "I.am."+temp1+"@gmail.com"
            
            emercon = Person()
            temp2 = self.names[r.randint(0,len(self.names)-1)]
            phone2 = r.randint(6000000000,9999999999)
            email2 = "Emergency."+temp2+"@gmail.com"
            rel = self.relation[r.randint(0,len(self.relation)-1)]
            emercon.addPerson(temp2,rel,phone2,email2)
            self.__patient[i].addContact(phone1,email1,None,emercon)

    def addPatient(self):
        new_id = len(self.__patient.keys())+1
        new=Patient()
        name = input("Enter Name: ")
        name = name.split(" ")
        dob = input("Enter DOB: ")
        gender = input("Enter Gender: ")
        blood = input("Enter Blood Group: ")
        contact = int(input("Enter Mobile Number: "))
        new.addPatient(name[0],name[1],gender,blood,dob)
        new.addContact(contact)
        self.__patient[new_id]=new

    def delPatient(self,pid):
        if pid not in self.__patient:
            raise("Patient Id not found")
        print(f"Deleting id: {pid}")
        print("-"*35)
        del self.__patient[pid]
        iterate = len(self.__patient.keys())
        for i in range(pid+1,iterate+2):
            self.__patient[i-1]=self.__patient[i]
        del self.__patient[iterate+1]

    def showPatient(self):
        iterate = len(self.__patient.keys())
        for i in range(1,iterate+1):
            print(f"ID: {i} Name: {self.__patient[i].returnAtt('firstname')} {self.__patient[i].returnAtt('lastname')}")
        print("-"*35)

    def showPatientById(self,pid,emergency=None):
        if pid not in self.__patient:
            raise("Patient Not Found")
        if emergency==None:
            self.__patient[pid].show()
        else:
            self.__patient[pid].returnAtt("emergency_contact").show()
        
    def updatePatient(self,pid):
        if pid not in self.__patient:
            print("Patient Not Found")
            return False
        temp = self.__patient[pid].attributes
        temp=temp[:-2]
        print("Available Attributes:")
        for i in range(len(temp)):
            print(f"{i+1}: {temp[i]}")
        att = int(input("Select Attribte Number: "))
        att-=1
        upd = input("Enter Details: ")
        #print(f"{temp[att]} updated from {self.__patient[pid].returnAtt(temp[att])} to {upd}")
        self.__patient[pid].update(temp[att],upd)
        
    def addDoctor(self):
        new_id = len(self.__doctor.keys())+1
        new=Doctor()
        name=input("Enter Name: ")
        name=name.split(" ")
        gender = input("Enter Gender: ")
        spec = input("Enter Specialization: ")
        contact = int(input("Enter Contact Number: "))
        email = input("Enter Email Id: ")
        new.addDoctor(True,name[0],name[1],gender,spec,contact,email)
        self.__doctor[new_id]=new

    def delDoctor(self,did):
        if did not in self.__doctor:
            raise("Doctor Id not found")
        print(f"Deleting id: {did}")
        print("-"*35)
        del self.__doctor[did]
        iterate = len(self.__doctor.keys())
        for i in range(did+1,iterate+2):
            self.__doctor[i-1]=self.__doctor[i]
        del self.__doctor[iterate+1]

    def showDoctor(self):
        iterate = len(self.__doctor.keys())
        for i in range(1,iterate+1):
            print(f"ID: {i} Name: {self.__doctor[i].returnAtt('firstname')} {self.__doctor[i].returnAtt('lastname')}")
        print("-"*35)

    def showDoctorById(self,did):
        if did not in self.__doctor:
            raise("Doctor Id not found")
        self.__doctor[did].show()
        
    def updateDoctor(self,did,sudo="None"):
        if did not in self.__doctor:
            raise("Doctor Not Found")
            
        temp = self.__doctor[did].attributes
        if sudo.lower()!="sudo":
            temp=temp[:-2]
        print("Available Attributes:")
        for i in range(len(temp)):
            print(f"{i+1}: {temp[i]}")
        att = int(input("Select Attribte Number: "))
        att-=1
        upd = input("Enter Details: ")
        print(f"{temp[att]} updated from {self.__doctor[did].returnAtt(temp[att])} to {upd}")
        self.__doctor[did].update(temp[att],upd)
        print("-"*35)
        
```

```python
import random as r # --- U know that current date thingy right,.... implement that here also bro ----
class Appointment():
    def __init__(self):
        self.attributes = ["AppID","DocID","PatID","AppDT","Status","VisitReason","VisitType","PriorityLevel","Notes","RoomNo"]
        self.data = {}
        for i in self.attributes:
            self.data[i] = None
        
    def addAppointment(self,appid=0,docid=0,patid=0,appdt="01/07/2007",stat="Not Attended",visitT="incabin",visitR="unhealthy",notes="Give this man some sleep!!!"):
        self.data["AppID"] = appid
        self.data["DocID"] = Hos.showDoctorById(docid)
        self.data["PatID"] = Hos.showPatientById(patid)
        self.data["AppDT"] = appdt
        self.data["Status"] = stat
        self.data["visitReason"] = visitR
        self.data["visitType"] = visitT
        self.data["PriorityLevel"] = round(1+3*r.random())
        self.data["Notes"] = notes
    
    def update(self,att,data):
        if att not in self.data:
            print(f"{att} not found in data")
            print("-"*35)
            return
        else:
            print(f"{att} changed from {self.data[att]} to {data}")
            print("-"*35)
            self.__data[att]=data
            
    def showAtt(self,att):
        if att not in self.data:
            print(f"{att} not found in data")
            print("-"*35)
            return
        else:
            print(f"{att} : {self.data[att]}")
            print("-"*35)
            
    def returnAtt(self,att):
        if att not in self.data:
            raise(f"{att} not found in data")
            return
        return self.data[att]
    
    def show(self):
        for i in self.attributes:
            print(i,":",self.data[i])
        print("-"*35)
```

Bill Class

```python
class Bill():
    def __init__(self):
        self.attributes = ["id","did","pid","billdt","stat","tamt","pamt","date","ins","mo_pay"]
        self.data = {}
        for i in self.attributes:
            self.data[i] = None
        
    def addBill(self,id=0,did=0,pid=0,billdt="01/07/2007",tamt=0,pamt=0,mo_pay="Enum"):
        self.data["id"] = id
        self.data["did"] = Hos.showDoctorById(did)
        self.data["pid"] = Hos.showPatientById(pid)
        self.data["billdt"] = billdt
        self.data["tamt"] = tamt
        self.data["pamt"] = pamt
        self.data["mo_pay"] = mo_pay
        if(not mo_pay):
            self.data["ins"] = True
        else:
            self.data["ins"] = False
        if(tamt==pamt):
            self.data["stat"] = "Paid"
        elif(pamt==0):
            self.data["stat"] = "Pending"
        else:
            self.data["stat"] = "Partial"
    
    def update(self,att,data):
        if att not in self.data:
            print(f"{att} not found in data")
            print("-"*35)
            return
        else:
            print(f"{att} changed from {self.data[att]} to {data}")
            print("-"*35)
            self.__data[att]=data
            
    def showAtt(self,att):
        if att not in self.data:
            print(f"{att} not found in data")
            print("-"*35)
            return
        else:
            print(f"{att} : {self.data[att]}")
            print("-"*35)
            
    def returnAtt(self,att):
        if att not in self.data:
            raise(f"{att} not found in data")
            return
        return self.data[att]
    
    def show(self):
        for i in self.attributes:
            print(i,":",self.data[i])
        print("-"*35)
```
